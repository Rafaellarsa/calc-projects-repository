```python
import json

with open("projects.json", "r") as f:
    projects = json.load(f)
```

```python
def extract_unique(projects, field):
    unique = set()
    for p in projects:
        for value in p.get(field, []):
            unique.add(value)
    return sorted(unique)
```

```python
fields = ["departments", "majors", "topics", "keywords"]

unique_dict = {field: extract_unique(projects, field) for field in fields}

unique_dict
```

```python
with open("searchFields.json", "w") as f:
    json.dump(unique_dict, f, indent=4)
```
